```python
# define the microlens

# =========
# distances
# =========

d_l = 4000  # distance of the lens in pc
d_s = 5000 # distance of the source in pc

# ===========
# lens system
# ===========

M0 = 0.05 # mass of the main lens in units of M_sol

M1 = 0.01  # secondary mass of the lens
d_01 = 0.01  # distance of M0 to M1 in units of AU

# ===========
# source size
# ===========

diameter_s = 1  
```

```python
#magnification is np.sum surface brightnesss times area of 
```

```python

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# compute lensing properties
from lenstronomy.Cosmo.micro_lensing import einstein_radius, source_size

theta_E = einstein_radius(M0, d_l, d_s)
print(theta_E, 'Einstein radius [arc seconds]')

size_s = source_size(diameter_s, d_s)
print(size_s, 'source size in arc seconds')
```

```python

# compute ray-tracing grid

grid_scale = size_s / 4.  # ten times the source size resolution
grid_width = theta_E * 4
num_pix = int(grid_width / grid_scale)
print(num_pix)

from lenstronomy.Util import util
x, y = util.make_grid(numPix=num_pix, deltapix=grid_scale)
```

```python
# set Einstein radius to zero - we just want to see the unlensed source for a moment
from lenstronomy.LensModel.lens_model import LensModel
lens = LensModel(lens_model_list=['POINT_MASS'])
kwargs_lens = [{'theta_E': 0, 'center_x': 0, 'center_y': 0}]
beta_x, beta_y = lens.ray_shooting(x, y, kwargs=kwargs_lens)

# compute surface brightness of the unlensed source
from lenstronomy.LightModel.light_model import LightModel
light = LightModel(light_model_list=['ELLIPSOID'])
kwargs_light = [{'amp': 1, 'radius': size_s/2 * 10, 'e1': 0, 'e2': 0, 'center_x': 0, 'center_y': 0}]

surface_brightness = light.surface_brightness(beta_x, beta_y, kwargs_light)
unlensed_brightness = np.sum(surface_brightness) # reference value when computing magnifications
```

```python
# map image to source plane
from lenstronomy.LensModel.lens_model import LensModel
lens = LensModel(lens_model_list=['POINT_MASS'])
kwargs_lens = [{'theta_E': theta_E, 'center_x': 0, 'center_y': 0}]

beta_x, beta_y = lens.ray_shooting(x, y, kwargs=kwargs_lens)

# compute surface brightness
from lenstronomy.LightModel.light_model import LightModel
ligth = LightModel(light_model_list=['ELLIPSOID'])
kwargs_light = [{'amp': 1, 'radius': size_s/2 * 10, 'e1': 0, 'e2': 0, 'center_x': 0.15*theta_E, 'center_y': 0}]

surface_brightness = ligth.surface_brightness(beta_x, beta_y, kwargs_light)
image = util.array2image(surface_brightness)
```

```python
# map image to source plane
from lenstronomy.LensModel.lens_model import LensModel
lens = LensModel(lens_model_list=['POINT_MASS'])
kwargs_lens = [{'theta_E': theta_E, 'center_x': 0, 'center_y': 0}]

beta_x, beta_y = lens.ray_shooting(x, y, kwargs=kwargs_lens)
save_results_to = '/Users/sameersingh/Desktop/Results/'
# compute surface brightness
from lenstronomy.LightModel.light_model import LightModel
ligth = LightModel(light_model_list=['ELLIPSOID'])
save_results_to = '/Users/sameersingh/Desktop/image/'
magnifications = []
multipliers = []
for val in range(-13,14):
    multiplier = val / 10
    kwargs_light = [{'amp': 1, 'radius': size_s/2 * 10, 'e1': 0, 'e2': 0, 'center_x': multiplier*theta_E, 'center_y': 0.1*theta_E}]
    surface_brightness = ligth.surface_brightness(beta_x, beta_y, kwargs_light)
    magnification = np.sum(surface_brightness)/unlensed_brightness
    magnifications.append(magnification)
    multipliers.append(multiplier)
    image = util.array2image(surface_brightness)
    print("This plot corresponds to x = " + str(multiplier) + " Einstein radii. The magnification is " + str(np.sum(surface_brightness)/unlensed_brightness))
    plt.imshow(image)
    plt.colorbar()
    plt.show()
```
